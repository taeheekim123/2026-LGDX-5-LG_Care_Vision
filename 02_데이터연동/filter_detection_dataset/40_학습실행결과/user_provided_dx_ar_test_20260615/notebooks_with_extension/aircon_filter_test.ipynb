####valid 이미지로 예측 확인

```python
import supervision as sv
import random

# 1. valid 데이터 먼저 로드
ds = sv.DetectionDataset.from_yolo(
    images_directory_path=f"{dataset.location}/valid/images",
    annotations_directory_path=f"{dataset.location}/valid/labels",
    data_yaml_path=f"{dataset.location}/data.yaml"
)

split_name = "valid"
print("검증 이미지 개수:", len(ds))

# 2. valid가 비어 있으면 train 데이터 사용
if len(ds) == 0:
    ds = sv.DetectionDataset.from_yolo(
        images_directory_path=f"{dataset.location}/train/images",
        annotations_directory_path=f"{dataset.location}/train/labels",
        data_yaml_path=f"{dataset.location}/data.yaml"
    )
    split_name = "train"
    print("학습 이미지 개수:", len(ds))

if len(ds) == 0:
    raise ValueError("valid/train 모두 불러온 이미지가 없습니다.")

# 3. 확인할 이미지 고정
i = 0
_, image, target = ds[i]

print("사용 데이터:", split_name)
print("사용 이미지 index:", i)
print("정답 라벨 개수:", len(target))

# 4. conf별 탐지 개수 확인
for conf in [0.03, 0.05, 0.1, 0.15, 0.2]:
    results = model(image, verbose=False, conf=conf, iou=0.4)[0]
    detections = sv.Detections.from_ultralytics(results).with_nms(threshold=0.4)
    print("conf:", conf, "탐지 개수:", len(detections))

# 5. 실제 표시할 conf 선택
selected_conf = 0.1

results = model(image, verbose=False, conf=selected_conf, iou=0.4)[0]
detections = sv.Detections.from_ultralytics(results).with_nms(threshold=0.4)

print("선택한 conf:", selected_conf)
print("최종 탐지 개수:", len(detections))

box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

image2 = image.copy()
image2 = box_annotator.annotate(scene=image2, detections=detections)
image2 = label_annotator.annotate(scene=image2, detections=detections)

sv.plot_image(image2)
```

```python
import supervision as sv
import matplotlib.pyplot as plt

if len(ds) == 0:
    raise ValueError("불러온 이미지가 없습니다.")

# 확인할 이미지 고정
i = 0
_, image, target = ds[i]

print("사용 이미지 index:", i)
print("정답 라벨 개수:", len(target))

# conf별 탐지 개수 확인
for conf in [0.03, 0.05, 0.1, 0.15, 0.2]:
    results = model(image, verbose=False, conf=conf, iou=0.4)[0]
    detections = sv.Detections.from_ultralytics(results).with_nms(threshold=0.4)
    print("conf:", conf, "탐지 개수:", len(detections))

# 실제로 화면에 표시할 conf 선택
selected_conf = 0.1

results = model(image, verbose=False, conf=selected_conf, iou=0.4)[0]
detections = sv.Detections.from_ultralytics(results).with_nms(threshold=0.4)

print("선택한 conf:", selected_conf)
print("최종 탐지 개수:", len(detections))

box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# 정답 라벨 이미지
gt_image = image.copy()
gt_image = box_annotator.annotate(scene=gt_image, detections=target)
gt_image = label_annotator.annotate(scene=gt_image, detections=target)

# 모델 예측 이미지
pred_image = image.copy()
pred_image = box_annotator.annotate(scene=pred_image, detections=detections)
pred_image = label_annotator.annotate(scene=pred_image, detections=detections)

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.title("Ground Truth")
plt.imshow(gt_image)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title(f"Prediction conf={selected_conf}")
plt.imshow(pred_image)
plt.axis("off")

plt.show()
```
